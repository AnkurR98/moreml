```python
from sklearn import tree
import pandas as pd
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Assets/pima-indians-diabetes-database/diabetes.csv')
```

```python
X = data.drop('Outcome', axis=1)
labels = X.columns.tolist()
y = data['Outcome']
```

```python
import os
os.environ["PATH"] += os.pathsep + 'D:/Program Files (x86)/Graphviz2.38/bin/'
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(X, y)
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=labels, class_names=['0', '1'], filled=True, rounded=True, special_characters=True)
graph= pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('diabetesinfogain.png')

clf = tree.DecisionTreeClassifier(criterion="gini")
clf = clf.fit(X, y)
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=labels, class_names=['0', '1'], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('diabetesgini.png')
```

```python
plt.figure(figsize=(100, 20))
im1 = plt.imread('diabetesinfogain.png')
plt.imshow(im1)

plt.figure(figsize=(100, 20))
im2 = plt.imread('diabetesgini.png')
plt.imshow(im2)
```

```python
data1 = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/639388c2cbc2120a14dcf466e85730eb8be498bb/iris.csv')
data1.columns
X = data1.drop('species', axis=1)
labels = X.columns.tolist()
y = data1['species']
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(X, y)
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=labels, class_names=['setosa','versicolor','virginica'], filled=True, rounded=True, special_characters=True)
graph= pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('irisinfogain.png')

clf = tree.DecisionTreeClassifier(criterion="gini")
clf = clf.fit(X, y)
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=labels, class_names=['setosa','versicolor','virginica'], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('irisgini.png')
```

```python
plt.figure(figsize=(100, 20))
im1 = plt.imread('irisinfogain.png')
plt.imshow(im1)

plt.figure(figsize=(100, 20))
im2 = plt.imread('irisgini.png')
plt.imshow(im2)
```
